```python
import os
import glob as glob
root_path = 'input/'
train_path = f'{root_path}hf_round1_train/train/'
test_path = f'{root_path}hf_round1_testA/testA/'
train_files = glob.glob(f'{train_path}*.txt')
test_files = glob.glob(f'{test_path}*.txt')
label_df = pd.read_csv(f'{root_path}lable.csv',index_col=0)
```

```python
BATCH_SIZE = 49
def load_data(case):#定义更多参数，具体参考官方说明
    df = pd.read_csv(case,sep=" ")
    df['III'] = df['II']-df['I']
    df['aVR']=-(df['II']+df['I'])/2
    df['aVL']=(df['I']-df['II'])/2
    df['aVF']=(df['II']-df['I'])/2
    return df 

from scipy.fftpack import fft,ifft
def myfft(ts,n=60):
    s = fft(ts)
    m = len(s)
    cutfun = np.ones([m,1])
    cutfun[20:m-20] = 0
    ss = s
    ss[n:m-n] = 0
    f = ifft(ss)
    real_f = np.real(f)
    
#     plt.plot(ts)
#     plt.plot(real_f)
#     plt.show()
    return real_f

def get_prepared_x(case,de_sample_freq = 20,is_fft=True):
    df = load_data(case)
    res_df = pd.DataFrame()
    if(is_fft):
        for col in df.columns:
            s = np.array(df[col])[[i for i in range(0,len(df),de_sample_freq)]]
            res_df[col] = myfft(s)
    else:
        res_df = df
    return np.array(res_df).transpose(1,0).reshape(-1)

def get_batch_data(batch_size=BATCH_SIZE,is_train=True,class_col=0,start=0):
    res_x = []
    res_y = []
    if is_train is True:
        length = len(train_files)
        ids = np.random.randint(0,length,size=batch_size,dtype=np.int32)
        for _id in ids:
            case = train_path + label_df.index.values[_id]
#             print(case.split('/')[-1])
            res_x.append(get_prepared_x(case))
            res_y.append(label_df[arrythmia_values[class_col]].values[_id])
#             print(label_df[arrythmia_values[class_col]].values[_id])
        return np.array(res_x),np.array(res_y)
    else:
        for i in range(batch_size):
            res_x.append(get_prepared_x(test_files[i + start]))
        next_start = start+batch_size
        return np.array(res_x),next_start
```

```python
from xgboost.sklearn import XGBClassifier
from sklearn.metrics import f1_score
res_df = pd.DataFrame()

def train(cf=XGBClassifier(),class_col=0):
    for epoch in range(1):
        print
        for i in range(len(train_files) // BATCH_SIZE):
            if i % 100 is 0:
                print(f'{epoch} epoch, {i}')
            x,y = get_batch_data(class_col=class_col)
#             print(x.shape,y.shape)
            cf.fit(x,y)
    res = []
    next_start = 0
    for i in range(len(test_files) // BATCH_SIZE):
        x,next_start = get_batch_data(is_train=False,start=next_start)
        y_hat = cf.predict(x)
        res.append(y_hat)
    return cf,res
```

```python
other_params = {'n_jobs':6}
cf = XGBClassifier(**other_params)
for class_col in range(23,55):
    _,res = train(cf=cf,class_col=0)
    res_df[class_col] = np.array(res).reshape(-1)
    res_df.to_csv('pre.csv')
```
